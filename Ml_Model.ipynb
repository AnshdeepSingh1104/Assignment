#### **Import Data**

```python
df = pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/MPG.csv')
```

```python
df.head()
```

```python
df.nunique()
```

#### **Describe Data**

```python
df.info()
```

```python
df.describe()
```

#### **Data Visualization**

```python
sns.pairplot(df, x_vars = ['displacement', 'horsepower', 'weight', 'acceleration', 'mpg'], y_vars = ['mpg']);
```

```python
sns.regplot(x = 'displacement',y = 'mpg',data = df);
```

#### **Data Preprocessing**

```python
df.corr()
```

```python
df=df.dropna()
```

```python
df.info()
```

#### **Define Target Variable (y) and Feature Variables (X)**

```python
df.columns
```

```python
y=df['mpg']
```

```python
y.shape
```

```python
x=df[['displacement','horsepower','weight','acceleration']]
```

```python
x.shape
```

```python
x
```

#### **Scaling Data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
```

```python
X=ss.fit_transform(x)
```

```python
X
```

```python
pd.DataFrame(X).describe()
```

###### **After Standardization Mean is Zero and Standard Deviation is One**

#### **Train Test Split Data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train , Y_test = train_test_split(X,y, train_size = 0.7, random_state = 2529)
```

```python
X_train.shape, X_test.shape, Y_train.shape , Y_test.shape
```

#### **Modeling**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

#### **Prediction**

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

#### **Model Evaluation**

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error ,r2_score
```

```python
mean_absolute_error(Y_test,y_pred)
```

```python
mean_absolute_percentage_error(Y_test, y_pred)
```

```python
r2_score(Y_test,y_pred)
```

### **Explanation**

In this project, we built a linear regression model to predict vehicle mileage (mpg) based on features such as displacement, horsepower, weight, and acceleration.

**Data Exploration:** We imported the dataset and explored it to understand its structure and contents.
Descriptive statistics and visualizations helped us understand the distributions and relationships among the variables.

**Data Preprocessing:** We handled missing values by dropping rows with missing data.
We standardized the feature variables to have a mean of 0 and a standard deviation of 1, which is essential for linear regression models.

**Model Building:** We split the data into training and testing sets (70% training, 30% testing) to evaluate the model's performance on unseen data.
We trained a linear regression model using the training data.

**Model Evaluation:** The model's intercept was approximately 23.49, and the coefficients for displacement, horsepower, weight, and acceleration were -1.06, -1.69, -4.11, and -0.11, respectively.
The Mean Absolute Error (MAE) was about 3.33, indicating the average prediction error in mpg.
The Mean Absolute Percentage Error (MAPE) was around 14.71%, showing the average percentage error.
The R-squared (R) score was 0.703, meaning the model explains 70.3% of the variance in mpg, indicating a good fit.

**Conclusion**

Our linear regression model provides a reasonable prediction of vehicle mileage based on the selected features. The R score of 0.703 suggests the model captures a significant portion of the variance in mileage. To further improve the model, we could explore additional features, try advanced regression techniques, or conduct more feature engineering.
